```python
# seach patterns
# non-breaking space
nbspace = r'\u00A0'
# more than one space character in a row
additionalSpaces = r'\s+'
# space before comma
spaceComma = r'\s,'
# comma after exclamation mark
exclamComma = r'!,'
# comma after question mark
questionComma = r'\?,'
# comma after full stop
fullStopComma = r'\.,'
# comma after comma
doubleComma = r',,'
# comma at beginning of text line
startComma = r'^,\s*'
# (, or (,) 
parenthesesComma = r'\(,+\)*'
# three or more full stops
additionalStops = r'\.\.\.+'

for text in lotr['dialog']:
    textString = str(text)
    textString = re.sub(nbspace, ' ', textString)# geschützte Leerzeichen durch normale Leerzeichen ersetzen
    textString = re.sub(additionalSpaces,' ', textString) # mehrfache Leerzeichen zu einfachen
    textString = textString.strip() # Leerzeichen am Zeilenanfang /-ende entfernen
    textString = re.sub(spaceComma,',', textString) # Leerzeichen vor Kommata weg
    textString = re.sub(exclamComma,'! ', textString) # Komma nach Ausrufezeichen weg
    textString = re.sub(questionComma,'? ', textString) # Komma nach Fragezeichen weg
    textString = re.sub(fullStopComma,'. ', textString) # Komma nach Punkt weg
    textString = re.sub(doubleComma,', ', textString) # doppelte Kommata weg
    textString = re.sub(startComma,'', textString)# Komma am Anfang des Strings löschen
    textString = re.sub(parenthesesComma, '', textString) # löschen von (,) oder (,
    textString = re.sub(additionalStops, '...', textString) # reduce ...... to ...
    
    # zur Sicherheit noch mal überflüssigen Whitespace löschen (beim Ersetzen der überflüssigen Kommata könne manchmal doppelte Leerzeichen entstehen, wenn schon ein Leerzeichen da war)
    textString = re.sub(additionalSpaces,' ', textString) # mehrfache Leerzeichen zu einfachen
    #print(textString)
    
    # how to write back to data frame? this does not work
    # text = textString
    
    #language detection with cleaned data
    language = detect(textString)
    #print(language)
    if (language != "en"):
        print(textString)
        print(language)
```
